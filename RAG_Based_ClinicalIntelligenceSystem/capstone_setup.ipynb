```python
import sys

def get_serverless_version() -> int:
  python_version = sys.version.split()[0]
  match python_version:
    case "3.10.12":
      return 1
    case "3.11.10":
      return 2
    # default to latest version
    case _:
      return 2

serverless_version = get_serverless_version()
print(serverless_version)
```
